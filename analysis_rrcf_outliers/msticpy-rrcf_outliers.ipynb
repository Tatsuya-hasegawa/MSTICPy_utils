# Using Robust Random Cut Forest with rrcf python module

````python
# Pre Processing
import re
import pandas as pd

def replace_column_name(column_name):
    column_name = re.sub(r'\((.*?)\)', lambda m: '_' + m.group(1), column_name)
    column_name = column_name.replace('.', '_')
    return column_name

rrcf_df = base_df.copy()

rrcf_df.set_index("_time", inplace=True)
rrcf_df.fillna(0, inplace=True)
for col in rrcf_df.columns:
    rrcf_df[col] = rrcf_df[col].astype(float)

rrcf_df = rrcf_df.drop("_span",axis=1)

new_columns = rrcf_df.columns.map(replace_column_name)
rrcf_df.columns = new_columns
print(rrcf_df.columns)

rrcf_df.head()
````

````python
from msticpy.analysis.outliers import identify_outliers_rrcf

# Feature selection
feature_columns = rrcf_df.columns.tolist()
print("Selected features:", feature_columns)
print("Number of features:", len(feature_columns))

#feature_columns = ['count','dc_src_ip']
feature_columns = ['count', 'dc_src_ip', 'dc_dest_ip', 'sum_flow_bytes_toclient', 'sum_flow_bytes_toserver', 'sum_flow_pkts_toclient', 'sum_flow_pkts_toserver']
X = rrcf_df[feature_columns].values

# Check the shape of X
print("Shape of X:", X.shape)
````

````python
# Perform anomaly detection
clf, X_outliers, y_pred_outliers = identify_outliers_rrcf(X, X, contamination=0.01,max_features=X.shape[1],max_samples=X.shape[0])

# Add outlier flag to the original DataFrame
rrcf_df['outlier'] = y_pred_outliers

# Display detected outliers
print("Detected outliers:")
display(rrcf_df[rrcf_df['outlier'] == -1])

# Display statistics of outliers
print("\nOutlier statistics:")
print(rrcf_df['outlier'].value_counts())

# Display feature-wise statistics for normal and outlier data
print("\nFeature statistics for normal and outlier data:")
for feature in feature_columns:
    print(f"\n{feature}:")
    print(rrcf_df.groupby('outlier')[feature].describe())
    
````

````python
## Time Series Graph
from msticpy.nbtools.timeseries import display_timeseries_anomolies

output = rrcf_df.copy()

# When using the display_timeseries_anomolies function, apply the datetime type to the time field of the result and sort it in chronological order in advance.
rrcf_df['date_clock'] = pd.to_datetime(rrcf_df.index)
rrcf_df = rrcf_df.sort_values(by='date_clock')
rrcf_df['baseline'] = rrcf_df['count']

# If the outlier column is -1, set the abnormal column to 1, otherwise set it to 0
rrcf_df['anomalies'] = (rrcf_df['outlier'] == -1).astype(int)

timeseries_anomalies_plot = display_timeseries_anomolies(
    data=rrcf_df, 
    y='count',
    time_column='date_clock'
    )

````
