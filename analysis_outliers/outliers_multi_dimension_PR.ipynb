```python
# 7 dimension 

import numpy as np
from msticpy.analysis.outliers import identify_outliers,plot_outlier_results

n_dimension = 7

# create random numeric samples
data = np.random.rand(100, n_dimension)

# calc outliers by Isolation Forest algorism
clf, X_outliers, y_pred_outliers = identify_outliers(data, data, contamination=0.1, max_features=0.4)
    
feature_columns = [f'feature{i}' for i in range(1, n_dimension+1)]

plot_outlier_results(
    clf, 
    data,
    data, 
    X_outliers, 
    feature_columns=feature_columns, 
    plt_title="MSTICPY Isolation Forest Anomaly Detection for Multi Dimension Features"
)
```

```python
# Choice 2 dimension from 6 features 

import numpy as np
import pandas as pd
from msticpy.analysis.outliers import identify_outliers,plot_outlier_results

n_dimension = 6

data = np.random.rand(100, n_dimension)
clf, X_outliers, y_pred_outliers = identify_outliers(data, data, contamination=0.1)
    
feature_columns = [f'feature{i}' for i in range(1, n_dimension+1)]

df = pd.DataFrame(data, columns=feature_columns)
chosen_feature_columns = ["feature3", "feature5"]

plot_outlier_results(
    clf, 
    df[['feature3', 'feature5']].values,
    df[['feature3', 'feature5']].values, 
    X_outliers, 
    feature_columns=chosen_feature_columns, 
    plt_title="Chosen 2 features: MSTICPY Isolation Forest Anomaly Detection for Multi Dimension Features"
)   
    
```
