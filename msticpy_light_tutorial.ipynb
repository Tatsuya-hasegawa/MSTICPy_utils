### Running a data query

To run a query you must load and instantiate a Query Provider for the data source that you want to query from. 

Several providers are supported by MSTICPy including: Microsoft Sentinel, Splunk, Microsoft Defender, Azure Monitor, Cybereason and Sumologic.

For example, `Splunk` is following. It get the data via 8089 (Splunk REST API ).

ref: https://msticpy.readthedocs.io/en/latest/data_acquisition/SplunkProvider.html

password: dummypassword

role: `splunk-system-role`  (just user or power role failed, probably admin or splunk-system-role)

````python
splunk_uri = "localhost"
splunk_user = "api_msticpy" # change to your true splunk user name
splunk_password = "dummypassword" # change to your true splunk the user's password
qry_prov = mp.QueryProvider("Splunk")
qry_prov.connect(host=splunk_uri, username=splunk_user, password=splunk_password) # user -> username
````

**Alternatively, you can read the Splunk credentinal info from config yaml file like this.**

````python
%%writefile msticpyconfig.yaml

DataProviders:
  Splunk:
      Args:
        host: localhost
        port: 8089
        username: "api_msticpy"  # change to your true splunk user name
        password: "dummypassword" # change to your true splunk the user's password
````

````python
splunk_prov = mp.QueryProvider("Splunk")
splunk_prov.connect()
````

**Running an ad hoc Splunk query**

````python
splunk_query = '''
search index="_internal"
| head 5
| table *
'''
df = splunk_prov.exec_query(splunk_query)
df.head()
````

````python
splunk_query = '''
search index="_internal" 
| head 100
| stats count by name,group
| sort - count
| table name group count
'''
df = splunk_prov.exec_query(splunk_query)
df.head()
````
